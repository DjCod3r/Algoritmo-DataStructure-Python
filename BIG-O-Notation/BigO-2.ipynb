### Funções Big-O

Constant | 
Log(n) Logarithmic | 
n Linear | 
nLog(n)log linear | 
n*2 Quadratic | 
n*3 Cubic | 
2^n Exponencial

```python
from math import log
import numpy as np
import matplotlib.pyplot as plt
```

```python
n = np.linspace(1, 10, 100); n #geração de numeros de 1 a 10
```

```python
labels = ['Constant', 'Logarithmic', 'Linear', 'Log Linear', 'Quadratic', 'Cubic' , 'Exponential']
big_o =  [np.ones(n.shape), np.log(n), n , n * np.log(n), n**2, n**3, 2**n]
```

```python
plt.figure(figsize=(10 , 8))
plt.ylim(0,100)
for i in range(len(big_o)):
    plt.plot(n, big_o[i], label = labels[i])
plt.legend()
plt.ylabel('Tempo de Execução')
plt.xlabel('Entradas (n)')
```

BIG-O | Função Constante - O(1)

```python
lista = [1,2,3,4,5,6,7,8,9]
def constante(n): #O = 1 contem 1 passo
    print(n[0])

constante(lista)
```

Funcão Linear - O(n)

```python
def linear(n): #O = n contem n passos
    for i in n:
        print(i)

linear(lista)
```

Função Quadratica - O(n^2)

```python
def quadratica(n): #O = n² contem n² passos
    for i in n:
        for j in n:
            print(i, j)

quadratica(lista)
```

Combinação

```python
def combinacao(n): #O(2) + O(n) = O(n)
    print(n[0])

    for i in range(5):
        print('Valor de i: ', i)
    
    for i in n:
        print(i)
    
    print('teste')

combinacao(lista)
```
